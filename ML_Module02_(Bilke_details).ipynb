```python
#resetting the index
df_copy = df.copy()

df_copy.reset_index(drop = True, inplace = True)
df_copy
```

```python
#1.What is the range of selling prices in the dataset?


min_price = df_copy['selling_price'].min()
max_price = df_copy['selling_price'].max()
price_range = max_price - min_price

print(f"Range of Selling Prices: {price_range}")
```

```python
#2.What is the median selling price for bikes in the dataset?

median_price = df_copy['selling_price'].median()
print(f"Median Selling Price: {median_price}")
```

```python
#3.What is the most common seller type?

seller_type_counts = df_copy['seller_type'].value_counts()
most_common_seller_type = seller_type_counts.idxmax()
print(f"Most Common Seller Type: {most_common_seller_type}")
```

```python
#4.How many bikes have driven more than 50,000 kilometers?

bikes_driven_more_than_50k = (df_copy['km_driven'] > 50000).sum()

print(f"Bikes driven more than 50,000 : {bikes_driven_more_than_50k}")
```

```python
#5.What is the average km_driven value for each ownership type?

average_km_driven_by_ownership = df_copy.groupby('owner')['km_driven'].mean()

print("Average km_driven value for each ownership type:")
print(average_km_driven_by_ownership)
```

```python
#6.What proportion of bikes are from the year 2015 or older?

bikes_from_recent_years = (df_copy['year'] >= 2015).sum()
total_bikes = len(df_copy)
proportion_recent_years = bikes_from_recent_years / total_bikes


print(f"Proportion of bikes from the year 2015 or older: {proportion_recent_years:.2f}%")
```

```python
#7.What is the trend of missing values across the dataset?

missing_values_trend = df_copy.isnull().sum() / len(df_copy)

print("Trend of missing values across the dataset:")
print(missing_values_trend)

#as we have already done data preparation
```

```python
#9.What is the total number of bikes listed by each seller type?

bike_counts_by_seller_type = df_copy['seller_type'].value_counts()

print("Total number of bikes listed by each seller type:")
print(bike_counts_by_seller_type)
```

```python
#8.What is the highest ex_showroom_price recorded, and for which bike?


# Convert 'ex_showroom_price' column to numeric
df_copy['ex_showroom_price'] = pd.to_numeric(df_copy['ex_showroom_price'], errors='coerce')

highest_ex_showroom_price = df_copy['ex_showroom_price'].max()
bike_with_highest_price = df_copy[df_copy['ex_showroom_price'] == highest_ex_showroom_price]

print(f"Highest ex_showroom_price: {highest_ex_showroom_price}")
print("Bike with highest price:")
print(bike_with_highest_price)
```

```python
#11. Identify and remove outliers in the km_driven column using the IQR method?
```

```python
#11.Identify and remove outliers in the km_driven column using the IQR method
Q1 = df_copy['km_driven'].quantile(0.25)
Q3 = df_copy['km_driven'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_copy = df_copy[(df_copy['km_driven'] >= lower_bound) & (df_copy['km_driven'] <= upper_bound)]
```

```python
#12.Perform a bivariate analysis to visualize the relationship between year and selling_price.


average_price_by_year = df_copy.groupby('year')['selling_price'].mean().reset_index()

# Create the line plot
sns.lineplot(x='year', y='selling_price', data=average_price_by_year)
plt.title('Average Selling Price Trend over Years')
plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.show()
```

```python
#13.What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?


# 1. Calculate the age of each bike
current_year = df_copy.Timestamp.now().year  # Get the current year
df_copy['age'] = current_year - df_copy['year']

# 2. Group by age and calculate the average selling price
average_price_by_age = df_copy.groupby('age')['selling_price'].mean().reset_index()

# 3. Calculate depreciation (assuming original price is the selling price at age 0)
original_price = average_price_by_age.loc[average_price_by_age['age'] == 0, 'selling_price'].values[0]
average_price_by_age['depreciation'] = original_price - average_price_by_age['selling_price']

# 4. Display the average depreciation by age
print(average_price_by_age[['age', 'depreciation']])




```

```python
# Which bike names are priced significantly above the average price for their manufacturing yea

# Calculate the average selling price for each year
average_price_by_year = df_copy.groupby('year')['selling_price'].mean()

# Function to identify significantly above-average priced bikes
def is_significantly_above_average(row):
    year = row['year']
    selling_price = row['selling_price']
    average_price = average_price_by_year.get(year, 0)  # Handle cases where a year might not have an average
    return selling_price > average_price * 1.2  # Consider prices 20% above average as significant

# Apply the function to each row
df_copy['significantly_above_average'] = df_copy.apply(is_significantly_above_average, axis=1)

# Get the names of bikes that meet the criteria
significantly_priced_bikes = df_copy[df_copy['significantly_above_average']]['name']

significantly_priced_bikes
```
