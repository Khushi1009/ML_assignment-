```python

#changing dtype of discounted_price to int dtype

def size_process(item):
    if str(item)[0] == "₹":
        try:
            res = float(str(item).replace("₹", ""))
            return res
        except ValueError:
            return np.NaN
    else:
        return np.NaN



df["discounted_price"] = df["discounted_price"].apply(size_process)
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors='coerce')
```

```python
df.info()
```

```python
 #convert actual_price to int

df['actual_price'] = df['actual_price'].astype(str)

def price_to_int(price_str):
    try:
        price_str = price_str.replace("₹", "").replace(",", "")
        return int(float(price_str))
    except (ValueError, AttributeError):
        return np.nan

df['actual_price'] = df['actual_price'].apply(price_to_int)
```

```python
# Convert 'rating' and 'rating_count' to int, handling errors
def to_int(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return np.nan

df['rating'] = df['rating'].apply(to_int)
df['rating_count'] = df['rating_count'].apply(to_int)
```

```python
#2.What are the top rating_count products by category?

df_sorted = df.sort_values(['category', 'rating_count'], ascending=[True, False])
top_n = df_sorted.groupby('category').head(3).reset_index(drop=True)

print(top_n[['category', 'product_name', 'rating_count']])
```

```python
#1.What is the average rating for each product category?
average_ratings = df.groupby('category')['rating'].mean().reset_index()
average_ratings = average_ratings.rename(columns={'rating': 'average_rating'})

print(average_ratings)
```

```python
#3.What is the distribution of discounted prices vs. actual prices?

plt.figure(figsize=(8, 6))
plt.scatter(df['discounted_price'], df['actual_price'], alpha=0.5)
plt.xlabel('Discounted Price')
plt.ylabel('Actual Price')
plt.title('Discounted vs. Actual Price Distribution')
plt.grid(True)


```

```python
#4.How does the average discount percentage vary across categories?

average_discount_percentage = df.groupby('category')['discount_percentage'].mean().reset_index()

```
